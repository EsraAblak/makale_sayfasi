```python
from transformers import pipeline
```

```python
# sentiment analysis

classifier = pipeline("sentiment-analysis")
```

```python
text = "It's great to learn NLP for me"
```

```python
classifier(text)
```

```python
import pandas as pd

outputs = classifier(text)
pd.DataFrame(outputs)
```

```python
# zero-shot classification

classifier = pipeline("zero-shot-classification")
```

```python
text = "This is a tutorial about Hugging Face"
```

```python
labels = ["tech", "education", "business"]
```

```python
outputs = classifier(text, labels)
pd.DataFrame(outputs)
```

```python
generator = pipeline("text-generation")
```

```python
prompt = "This tutorial will walk you through how to"
```

```python
outputs = generator(prompt)
```

```python
print(outputs)
```

```python
outputs = generator(prompt, max_length=50)
print(outputs[0]["generated_text"])
```

```python
generator = pipeline("text-generation",model="distilgpt2")
```

```python
outputs = generator(prompt, max_length=50)
print(outputs[0]["generated_text"])
```

```python
# name entity recognition

ner = pipeline("ner", grouped_entities=True)
```

```python
text = "My name is Esra from Turkey. HuggingFace is a nice platform."
```

```python
outputs = ner(text)
pd.DataFrame(outputs)
```

```python
text = "FirmX's customer service is fast and effective, FirmY's product quality is very poor"
```

```python
outputs = ner(text)
pd.DataFrame(outputs)
```

```python
# Q-A

reader = pipeline("question-answering")
```

```python
text = "My name is Esra and I love Berlin"
question = "Where do i like"
```

```python
outputs = reader(question=question, context=text)
pd.DataFrame([outputs])
```
